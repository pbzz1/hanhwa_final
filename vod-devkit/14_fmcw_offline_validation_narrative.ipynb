# FMCW 레이더 단독(최종) vs LiDAR 보조 검증(오프라인) — 발표·보고용 요약

이 노트북은 **코드 실행 없이** 구조와 말할 거리만 정리합니다. 상세 비교 실험은 `13_radar_bev_lidar_yolo_compare.ipynb`(B 구간)를 사용하세요.


## A. 최종 시스템 (운용·웹·발표의 메인 스토리)

**입력**: FMCW 레이더 단독  

1. **전처리**  
2. **군집화** (예: DBSCAN)  
3. **후보 선택**  
4. **추적**  
5. **위험도 계산**  

LiDAR는 **실제 배치 파이프라인의 입력으로 넣지 않는다**고 가정합니다.


## B. 오프라인 실험·검증 (데이터셋·노트북)

- 동일 후보(또는 동일 프레임)에 대해 **LiDAR·GT**로 **교차 검증**  
- **레이더-only 방식의 한계** 확인  
- **threshold·후처리 튜닝** 참고  

보고서·노트북에는 LiDAR를 넣어도 되지만, **「오프라인 비교용 / 시스템 입력 아님」**을 한 줄이라도 밝히면 논리가 깔끔합니다.


## 발표에서 안전한 표현 (복사용)

> 최종 운용 가정은 **FMCW 레이더 단독**이며, LiDAR는 실제 시스템 입력이 아니라 **데이터셋 기반 오프라인 검증용 보조 센서**로만 활용했습니다.

> LiDAR 검증은 레이더 후보의 **공간적 타당성**을 확인하기 위한 **실험적 비교 단계**이며, **실제 배치 파이프라인에서는 제외**됩니다.


## `strong_validated` / `radar_only` → 운영에서는 이렇게

**최종 FMCW-only UI**에서는 LiDAR 검증 여부로 후보를 나누는 라벨을 **메인 구분으로 두지 않는 것**이 좋습니다.

대신 **레이더 내부 정보**로 신뢰도 등급을 둡니다.

| 운영 등급 (예) | 의미 |
|----------------|------|
| `high_confidence` | 다수 프레임·안정적 특징 |
| `medium_confidence` | 일부 불안정 |
| `low_confidence` | 단편·불안정 |

**등급에 활용할 수 있는 레이더 특징** (LiDAR 불필요)

- 군집 점 수  
- RCS 안정성  
- 속도 일관성  
- 프레임 간 추적 지속성  
- heading 변화 안정성  

`strong_validated` 같은 이름은 **오프라인 분석·로그**에만 두고, 슬라이드·운영 화면에는 **위 등급 체계**로 맞추면 말이 통일됩니다.
