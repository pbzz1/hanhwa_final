# FMCW 레이더 단독 데모

노트북은 **`radar-service` 패키지를 import**해 한 프레임을 돌리는 용도만 둡니다. 알고리즘 본문은 `../app/` 모듈을 편집하세요.

```python
import sys
from pathlib import Path

# 노트북 cwd가 notebooks/일 때 상위를 패스에 추가
ROOT = Path.cwd().resolve().parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import numpy as np
from app.pipeline import run_pipeline_frame
from app.export_json import save_web_payload

# 합성 한 클러스터 (실데이터 대신: 아래 path를 VoD .bin으로 바꿀 것)
rng = np.random.default_rng(0)
xyz = rng.normal(size=(45, 3)) * 0.2 + np.array([20.0, 2.0, 0.0])
n = xyz.shape[0]
radar = np.column_stack([
    xyz.astype(np.float32),
    (np.abs(rng.normal(size=n)) * 5 + 5).astype(np.float32),
    rng.normal(size=n).astype(np.float32) * 0.2,
    rng.normal(size=n).astype(np.float32) * 0.4,
    np.zeros(n, dtype=np.float32),
])

payload, _ = run_pipeline_frame(radar, frame_id="demo", radar_frame_mode="radar")
out = ROOT / "outputs" / "web_payload.json"
save_web_payload(out, payload)
print("wrote", out)
print("detections:", len(payload.detections), "tracks:", len(payload.tracks))
```
